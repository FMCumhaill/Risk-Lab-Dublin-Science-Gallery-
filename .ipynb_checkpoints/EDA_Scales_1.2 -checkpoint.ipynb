```r
Data <- read.csv(file="RISKLAB_12_11_17.csv", header=TRUE, sep=",")
```

```r
names(Data)
```

## BIS Scale

```r
#Reverse code

library(car)

Data[[68]] <- recode(Data[[68]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")
Data[[74]] <- recode(Data[[74]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")
Data[[75]] <- recode(Data[[75]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")
Data[[76]] <- recode(Data[[76]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")
Data[[77]] <- recode(Data[[77]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")
Data[[79]] <- recode(Data[[79]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")
Data[[80]] <- recode(Data[[80]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")
Data[[82]] <- recode(Data[[82]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")
Data[[87]] <- recode(Data[[87]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")
Data[[96]] <- recode(Data[[96]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")
Data[[97]] <- recode(Data[[97]], "1 = 4 ; 2 = 3 ; 3 = 2 ; 4 = 1")

```

```r
#Calcaute sum values for total BIS scores 
Data["BIS_Sum"] <- 0
Data$BIS_Sum <- rowSums( Data[,68:97] )
```

```r
#Create a new column and adds up sums of other columns. Uses the " rowSums" function 

Data["BIS_Attention"] <- 0
Data$BIS_Attention <- rowSums(Data[,c(72,76,78,87,95)])

Data["BIS_Cog_Instability"] <- 0
Data$BIS_Cog_Instability <- rowSums(Data[,c(73,91,93)])

Data["BIS_Motor"] <- 0
Data$BIS_Motor <- rowSums(Data[,c(69,70,71,84,86,89,92)])

Data["BIS_Perseverance"] <- 0
Data$BIS_Perseverance <- rowSums(Data[,c(83,88,90,97)])

Data["BIS_self_control"] <- 0
Data$BIS_self_control <- rowSums(Data[,c(68,74,75,79,80,81)])

Data["BIS_Cog_Complexity"] <- 0
Data$BIS_Cog_Complexity <- rowSums(Data[,c(77,82,85,94,96)])
```

## DOSPERT Scale ##

```r
#Create a new column and adds up sums of other columns. Uses the " rowSums" function 

Data["DOS_Ethical"] <- 0
Data$DOS_Ethical <- rowSums(Data[,c(43,46,47,53,66,67)])

Data["DOS_Fin_Investment"] <- 0
Data$DOS_Fin_Investment <- rowSums(Data[,c(49,41,55)])

Data["DOS_Fin_Gambling"] <- 0
Data$DOS_Fin_Gambling <- rowSums(Data[,c(40,51,45)])

Data["DOS_HealthSafety"] <- 0
Data$DOS_HealthSafety <- rowSums(Data[,c(42,52,54,57,60,63)])

Data["DOS_Recreational"] <- 0
Data$DOS_Recreational <- rowSums(Data[,c(39,48,50,56,61,62)])

Data["DOS_Social"] <- 0
Data$DOS_Social <- rowSums(Data[,c(38,44,58,59,64,65)])
```
